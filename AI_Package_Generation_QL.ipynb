```python
user_input = "Kids"
budget = 2000

i = 0

while (i<5):
    package = {}
    for cl, cat in enumerate(df_category_elements.columns):
        if cat in user_input or user_input in cat:
            # print(cat)
            cat_ele = df_category_elements.loc[:, cat]
            idx = np.where(df_category_elements[cat].isna())[0][0]
            b1 = budget/5
            random_number = random.randint(0, idx)
            count = 0
            while((df_category_elements.iloc[random_number, cl+1] < b1)and (count < 10)):
                random_number = random.randint(0, idx)
                count += 1
            b1 = df_category_elements.iloc[random_number, cl+1]
            ele = cat_ele[random_number]
            package["Category Elements"] = ele

    for cl, cat in enumerate(df_category_theme.columns):
        if cat in user_input or user_input in cat:
            # print(cat)
            cat_theme = df_category_theme.loc[:, cat]

            if len(np.where(df_category_theme[cat].isna())[0]) == 0:
                idx = len(df_category_theme)-1
            else:
                idx = np.where(df_category_theme[cat].isna())[0][0]
            b2 = budget/5
            random_number = random.randint(0, idx)
            count = 0
            while(df_category_theme.iloc[random_number, cl+1] < b2 and count < 10):
                random_number = random.randint(0, idx)
                count += 1
            print(random_number, cl+1)
            b2 = df_category_theme.iloc[random_number, cl+1]
            ele = cat_theme[random_number]
            package["Theme"] = ele

    for i in range(len(df_theme_feature)):
        if package["Theme"] in df_theme_feature.iloc[i,1]:
            package["About Theme"] = df_theme_feature.iloc[i,1]
            break

    food_list = ""
    b3 = 0
    idx_b = random.choice(idx_beverage)
    idx_s = random.choice(idx_starter)
    idx_mc = random.choice(idx_main_course)
    idx_d = random.choice(idx_dessert)

    beverage = df_food.loc[idx_b, 'name']
    starter = df_food.loc[idx_s, 'name']
    main_course = df_food.loc[idx_mc, 'name']
    dessert = df_food.loc[idx_d, 'name']

    b3 += df_food.loc[idx_b, 'price']
    b3 += df_food.loc[idx_s, 'price']
    b3 += df_food.loc[idx_mc, 'price']
    b3 += df_food.loc[idx_d, 'price']

    package["Food"] = beverage + " " +starter + " " +main_course + " " +dessert
    package["Price"] = (b1 + b2 + b3)*(6/5)
    print(package)
    i += 1

# idx_beverage = [i for i, item in enumerate(df_food['category']) if item == "Beverage"]
# idx_starter = [i for i, item in enumerate(df_food['category']) if item == "Starter"]
# idx_main_course = [i for i, item in enumerate(df_food['category']) if item == "Main Course"]
# idx_dessert = [i for i, item in enumerate(df_food['category']) if item == "Dessert"]
```
